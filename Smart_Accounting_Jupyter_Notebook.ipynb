```python
# Read data from file 'filename.csv' 
# Control delimiters, rows, column names with read_csv
# From postfinance, the last two lines of the csv must be removed otherwise the dataframe cannot be imported.
data = pd.read_csv("export_Mouvements_20190824.csv", sep=';', header = 3, encoding='latin-1') 
# Preview the first 5 lines of the loaded data 
data.head()
```

```python
# Create the final dataframe with all the categories, for the moment contaning only values of zeros.
final_data = pd.DataFrame({'Date':[0], 'Text':[0], 'Credit':[0], 'Debit':[0], 'Valeur':[0], 'Solde':[0], \
            'Logement':[0], 'ImpotsTaxes':[0], 'Insurance':[0], 'Transport':[0], 'Telephonie':[0], 'Abonements':[0],\
             'Nourriture':[0], 'Diversification':[0], 'Pharma':[0], 'Autre':[0], 'Misc':[0]}) 
# Print the dataframe 
print(final_data) 
```
